```python
import sys
import os

sys.path.append(os.path.abspath("../")) 

from data import Tokenizer, Vocabulary, TextLoader, RNNTrainer
from models import DeepRNN, RNNLM
from utils import get_device
import torch
from torch import nn
import torch.nn.functional as F
import numpy as np
```

```python
filename = "../books/time_machine.txt"
data = TextLoader(filename, batch_size=1024, num_steps=32)
```

```python
rnns = DeepRNN(input_size=data.vocab_size, hidden_size=128, num_layers=3).to(device=get_device())

model = RNNLM(
    rnn=rnns,
    vocab_size=data.vocab_size,
).to(device=get_device())

model
```

```python
trainer = RNNTrainer(
    model=model,
    data=data,
    loss_name="cross_entropy",
    optimizer_name="adam",
    device=get_device(),
    lr=0.001,
    optimizer_params={},
    scheduler_name="reduce_on_platue"
)

trainer.train(max_epochs=125)
```
